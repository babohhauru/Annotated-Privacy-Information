```python
mycursor = mydb.cursor()

sql = '''CREATE TABLE IF NOT EXISTS ios_app_data (
    id int primary key auto_increment,
    app_name text,
    data_used_to_track_user text,
    data_linked_to_user text,
    data_not_linked_to_user text,
    data_not_collected text,
    no_details_provided text
)
'''

mycursor.execute(sql)
mydb.commit()
```

```python
def findName(soup):
    global app_name
    for x in soup.findAll('h1',class_='product-header__title app-header__title'):
        for span in soup.findAll('span',class_='badge badge--product-title'):
            span.decompose()
        app_name = x.text
        print(app_name)
    return app_name
```

```python
with open('ios.txt','r',encoding='utf-8') as f:
    urllst = f.readlines()
   # print(urllst)
       
for urls in urllst :
    page = requests.get(urls)
    soup = BeautifulSoup(page.text, "html.parser")
    
    app_name  = findName(soup)
    
    data_used_to_track_user,data_linked_to_user, data_not_linked_to_user,data_not_collected, no_details_provided = '','','','',''
    tags = soup.find_all(class_ = 'app-privacy__card')
    for tag in tags:
        #print(tag.h3)
        terms = tag.find_all('span',class_='privacy-type__grid-content privacy-type__data-category-heading')
        #print(tag.h3.text)
        for term in terms:
            #print(term.text)
            #print(tag.h3.text)
            if tag.h3.text == 'Data Used to Track You' :  
                data_used_to_track_user += term.text
                data_used_to_track_user += ";"
            elif tag.h3.text == 'Data Linked to You' :
                data_linked_to_user += term.text
                data_linked_to_user += ";"
            elif tag.h3.text == 'Data Not Linked to You' :
                data_not_linked_to_user += term.text
                data_not_linked_to_user += ";"
            elif tag.h3.text == 'Data Not Collected' :
                data_not_collected += term.text
                data_not_collected += ";"
            elif tag.h3.text == 'No Details Provided' :
                no_details_provided += term.text
                no_details_provided += ";"
            else:
                print("\n")
    
    mycursor = mydb.cursor()

    sql = '''INSERT INTO ios_app_data (
        app_name text,
        data_used_to_track_user text,
        data_linked_to_user text,
        data_not_linked_to_user text,
        data_not_collected text,
        no_details_provided text
    )
    '''
    sql = "INSERT INTO ios_app_data (app_name,data_used_to_track_user,data_linked_to_user,data_not_linked_to_user, data_not_collected,no_details_provided ) VALUES (%s, %s, %s, %s, %s, %s)"
    val = (app_name, data_used_to_track_user, data_linked_to_user, data_not_linked_to_user, data_not_collected, no_details_provided)
    mycursor.execute(sql, val)
    mydb.commit()
    
```
