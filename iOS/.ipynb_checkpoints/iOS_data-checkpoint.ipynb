```python
from bs4 import BeautifulSoup
import requests, time, random
import mysql.connector
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import threading
```

```python
# Connect to the MySQL database
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="violet",
  database="appinfo"
)
```

```python
# Create the cursor
mycursor = mydb.cursor()

# Create the 'ios_data' table if it doesn't exist
sql = '''CREATE TABLE IF NOT EXISTS ios_data (
    id int primary key auto_increment,
    app_name text,
    browsing_history text,
    contact_info text,
    contacts text,
    diagnostics text,
    financial_info text,
    health_and_fitness text,
    identifiers text,
    location text,
    other_data text,
    purchases text,
    search_history text,
    sensitive_info text,
    usage_data text,
    user_content text
)
'''

mycursor.execute(sql)
mydb.commit()

data_categories = [
    "Browsing History", "Contact Info", "Contacts", "Diagnostics", 
    "Financial Info", "Health & Fitness", "Identifiers", "Location", 
    "Other Data", "Purchases", "Search History", "Sensitive Info", "Usage Data", "User Content"
]
#data_categories = {category: set() for category in data_categories}
```

```python
def findName(soup):
    global app_name
    for x in soup.findAll('h1',class_='product-header__title app-header__title'):
        for span in soup.findAll('span',class_='badge badge--product-title'):
            span.decompose()
        app_name = x.text
        print(app_name)
    return app_name
```

```python
def scrape_data(url, data_categories):
    lock = threading.Lock()
    try:
        options = Options()
        options.binary_location = r'C:\Program Files\Mozilla Firefox\firefox.exe'
        options.add_argument('--headless')
        driver = webdriver.Firefox(executable_path='F:\python\geckodriver.exe', options=options)
        driver.get(url)
        
        #time.sleep(random.randint(1, 2))
        see_details_buttons = WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.XPATH, "//*[text()='See Details']"))
        )
        

        lock.acquire()
        for button in see_details_buttons: 
            button.click()
            
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        app_name  = findName(soup)
        
        h4_tags = WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.XPATH, "//h4[@class='privacy-type__data-category-heading']"))
        )
        
        data_categories = {category: set() for category in data_categories}
        for tag in h4_tags:
            #print(tag.text)
            category = tag.text
            ul_tags = tag.find_element(By.XPATH, "./following-sibling::ul[@class='privacy-type__category-items']")
            
            for li in ul_tags.find_elements(By.TAG_NAME, "li"):
                text = li.text
                data_categories[category].add(text)
                #print(text)
        mycursor = mydb.cursor()
        
        data = {}
        for category, data_set in data_categories.items():
            data[category] = ','.join(data_set)
        
        mycursor.execute("INSERT INTO ios_data (app_name, browsing_history, contact_info, contacts, diagnostics, financial_info, health_and_fitness, identifiers, location, other_data, purchases, search_history, sensitive_info, usage_data, user_content) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)", (app_name, data.get("Browsing History",""),data.get("Contact Info",""),data.get("Contacts",""),data.get("Diagnostics",""),data.get("Financial Info",""),data.get("Health & Fitness",""),data.get("Identifiers",""),data.get("Location",""),data.get("Other Data",""),data.get("Purchases",""),data.get("Search History",""),data.get("Sensitive Info",""),data.get("Usage Data",""),data.get("User Content","")))
        mydb.commit()
        
    except TimeoutException:
        print(f"Timeout exception occurred while loading {url}")
        
    except NoSuchElementException:
        print(f"Element not found exception occurred while loading {url}")
        
    finally:
        driver.quit()
```

```python
with open('ios.txt','r',encoding='utf-8') as f:
    urllst = f.readlines()

threads = []
for urls in urllst:
    thread = threading.Thread(target=scrape_data, args=(urls,data_categories))
    thread.start()
    threads.append(thread)

for thread in threads:
    thread.join()
```
